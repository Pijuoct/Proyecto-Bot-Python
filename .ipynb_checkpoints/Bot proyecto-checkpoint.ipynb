```python
from telegram.ext import Updater, CommandHandler, ConversationHandler, MessageHandler,Filters
from telegram import ChatAction

#===========================================================
# #[Opcional] Recomendable poner un log con los errores que apareceran por pantalla.
# import logging
# logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',level=logging.INFO)
# logger = logging.getLogger(__name__)
# def error_callback(update, context):
#     logger.warning('Update "%s" caused error "%s"', update, context.error)
#===========================================================


#telegram es la libreria instalda (pin install python-telegram-bot)

#updater recibe las actualziaciones de mensajes enviadas por el usuario o clic de botones
#concectamos el bot con telegram via al token proporcionado por el BotFather

#el CommandHandler es para manipular comandos

#el ConversatioHandler es para ejecutar codigos especificos que obtienen datos de entrada y salida distintos conforme
#el usuario interactua con el bot (conversa)

#MessageHandler aydua trabajar con los mensajes recibidos y enviados

#Filters es para filtrar los datos obtenidos

#ChatAction es para mostrar que acciones esta ejecutando el bot





#Comandos del bot:

#======================= 1RO "PREGUNTA INICIAL" ====================================
pregunta_inicial=0  #estado de la conversacion del bot al ejecutar /start

def start(update,context):
    
    chat=update.message.chat #obtiene la informacion de codigo del chat
    chat.send_action(
        action=ChatAction.TYPING,
        timeout=None
    )
    #esto mostrara que el bot esta escribiendo hasta que pase un tiempo predeterminado
    
    update.message.reply_text('Hola üòã, ¬øComo estas?')
    
    return pregunta_inicial

#esto significa que una vez el bot responda el /start se iniciara el estado 0 de la conversacion
#el conversationhandler leera el estado y ejecutara la funcion correspondiente a ese estado
    
    
    
#========================= 2DO "EVULAR ESTADO DE ANIMO" ==================================
def evaluar_animo(update,context):
    
    chat=update.message.chat #obtiene la informacion de codigo del chat
    chat.send_action(
        
        action=ChatAction.TYPING,
        timeout=None
    )
    #esto mostrara que el bot esta escribiendo hasta que pase un tiempo predeterminado
    
    
#     if(update.message.text.upper().find('BIEN')>0):
#         update.message.reply_text('Me alegro de leer eso üòÑ')
        
#     elif(update.message.text.upper().find('MAL')>0):
#         update.message.reply_text('Lamento oir eso, espero las cosas mejoren ü•∫üòì')
        
#     else:
#         update.message.reply_text('Lamento mi conocimiento limitado pero como catalogas ese estado de animo? üòÖ')
    
    
    
    
    lista(update,context)
    
    
    return ConversationHandler.END #cierra la conversacion

    
#========================= 3RO "LISTA CODIGO" ==================================
def lista(update,context):

    """Muestra la lista de comandos disponibles del bot"""
    
    chat=update.message.chat #obtiene la informacion de codigo del chat
    chat.send_action(
        action=ChatAction.TYPING,
        timeout=None
    )
    
    update.message.reply_text('Aqui estan la lista de Comandos que puedes usar üìñ:\n\n/list te volvera a enviar este mensaje con todos los comandos ‚úçÔ∏è\n/lyrics para buscar la letra de una canci√≥n üé§\n/singer para ir a ver la informacion y discografia de un artista üí´')
    
    
    
    
#===========================================================

#definicion del bloque de ejecucion princial del bot 

if __name__ == "__main__":
    
    #token del bot:   5329090044:AAEMLbE4fQfNxgUrcAoNXqrakXG5nMGG9Y4
    #para la API de telegram se debe pasar el use_context=true 
    
    updater = Updater(token='5329090044:AAEMLbE4fQfNxgUrcAoNXqrakXG5nMGG9Y4', use_context=True)
    
    dp= updater.dispatcher
    
    #dispatcher se encarga de enviar las acciones, al entrar un comando o un msm, se envia al dispatcher para ser manejado
    

#===========================================================
#     dp.add_handler(CommandHandler('start',start))
#     #el updater.dispatcher tendra una respuesta cuando se ingrese el comando start (texto de comillas), 
#     #y esa sera ejecutar la funcion start(segunda entrada del CommandHandler) 
#===========================================================

    dp.add_handler(CommandHandler('list',lista))

    dp.add_handler(ConversationHandler(
    
        entry_points=[
            #comandos a ejecutar cuando se llama al comando que inicia la conversacion
            CommandHandler('start',start)
            
        ],#puntos de entrada para iniciar la conversacion (comandos especificos),
        
        states={
            pregunta_inicial:[MessageHandler(Filters.text,evaluar_animo)]
            #se ejecutara un manipulador del mensaje y filtrara solo el texto, y sera llamada una respuesta que es una
            #funcion llamada evaluar_animo
            
            
        },#comandos a ejecutar dados los estados de la conversacion
        
        fallbacks=[]
        
    ))#cierra la entradas del conversationhandler
    
    updater.start_polling()
    updater.idle()
    
    #esto es lo que realmente activa el bot, crea un ciclo infinito en el cual siempre se esta atento a si el usuario envia
    #comandos nuevos
    
    
```
